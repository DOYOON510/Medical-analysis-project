```python
with open('df_all.pickle', 'rb') as f:
    df_all = pickle.load(f)
```

# Boxplot

```python
df_all['type'] = 'ori'
mca_df_all['type'] = 'mca'

df_merge = pd.concat([df_all, mca_df_all])
df_merge.reset_index(drop=True)
```

```python
feature1 = ['precision', 'recall', 'F-score', 'AUC']

for i in range(4):
    plt.figure(figsize=(10, 8), dpi=200)
    sns.set(font_scale=1.3)
    ax = sns.boxplot(data=df_merge,
                     x='model',
                     y=feature1[i],
                     order=['randomFst', 'adaBST', 'XGBoost'],
                     hue='type',
                     hue_order=['ori', 'mca'],
                     palette='pastel')

    ax.set_title(feature1[i], fontsize=25)
```
